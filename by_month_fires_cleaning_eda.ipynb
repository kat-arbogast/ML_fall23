```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Per Month Wildfire Data
us_fires_burn_monthly = pd.read_csv("./CleanData/us_fires_burn_monthly.csv")
us_fires_burn_monthly.info()
```

```python
my_fig = sns.barplot(data=us_fires_burn_monthly, x="Year", y="Acres Burned", errorbar=None, orient="v", color = "lightblue")
plt.xticks(rotation=45) 
```

```python
sns.barplot(data=us_fires_burn_monthly, x="Year", y="Number of Fires", errorbar=None, orient="v", color = "lightblue")
plt.xticks(rotation=45) 
```

```python
sns.barplot(data=us_fires_burn_monthly, x="Year", y="Acres Burned per Fire", errorbar=None, orient="v", color = "lightblue")
plt.xticks(rotation=45) 
```

```python
sns.scatterplot(data=us_fires_burn_monthly, x="Number of Fires", y = "Acres Burned per Fire", hue = "Month")
```

```python
sns.lineplot(data=us_fires_burn_monthly, x="Year", y="Acres Burned", errorbar = None)
```

```python
cols_list = ["Number of Fires", "Acres Burned per Fire", "Acres Burned"]

for col in cols_list:
    
    plt.figure(figsize=(20,5))
    
    my_fig = sns.catplot(data = us_fires_burn_monthly,
                         kind = "bar",
                         x = "Year",
                         y = col,
                         hue = "Month",
                         height = 6,
                         aspect = 5)     
    
    my_fig.set(title = f"{col} in the US for years 2000 to 2022")
    
    plt.savefig(f"./CreatedVisuals/{col}_2000_2022.png", dpi = 500)
    
```
